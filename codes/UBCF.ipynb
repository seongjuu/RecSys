```python
import pandas as pd
import numpy as np

# Read rating data
ratings = pd.read_csv('/home/sjkim/recommendSystem/finalproject/Amazon_ratings.csv', encoding='latin-1')
ratings.drop('timestamp', axis=1, inplace=True)
ratings
```

```python
ratings['rating'].mean()
```

```python
# Rating 데이터를 test, train으로 나누고 train을 full matrix로 변환
from sklearn.model_selection import train_test_split
x = ratings.copy()
y = ratings['user_id']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=12)
rating_matrix = x_train.pivot(values='rating', index='user_id', columns='item_id')        
```

```python
rating_matrix
```

```python
# 모든 가능한 사용자 pair의 Cosine similarities 계산
from sklearn.metrics.pairwise import cosine_similarity
matrix_dummy = rating_matrix.copy().fillna(0)
user_similarity = cosine_similarity(matrix_dummy, matrix_dummy)
user_similarity = pd.DataFrame(user_similarity, index=rating_matrix.index, columns=rating_matrix.index)
user_similarity     
```

```python
# 모든 user의 rating 평균 계산 
rating_mean = rating_matrix.mean(axis=1) # axis=1은 유저별 평균 
```

```python
# RMSE 계산을 위한 함수
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

def score(model):
    id_pairs = zip(x_test['user_id'], x_test['item_id'])
    y_pred = np.array([model(user, item) for (user, item) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

## UBCF_bias

```python
# 사용자의 평가경향을 고려한 추천
def ubcf_bias(user_id, item_id):
    import numpy as np
    user_mean = rating_mean[user_id]                                       # 현 user의 평점평균 가져오기
    if item_id in rating_matrix:
        sim_scores = user_similarity[user_id]                              # 현 user와 다른 사용자의 유사도 가져오기
        item_ratings = rating_matrix[item_id]                              # 현 movie의 모든 rating 가져오기
        others_mean = rating_mean                                          # 모든 사용자의 rating 평균 가져오기, 943개
        none_rating_idx = item_ratings[item_ratings.isnull()].index        # 현 movie에 대한 rating이 없는 user 삭제하려고 index 가져오기
        item_ratings = item_ratings.drop(none_rating_idx)
        sim_scores = sim_scores.drop(none_rating_idx)
        others_mean = others_mean.drop(none_rating_idx)
        item_ratings = item_ratings - others_mean                          # 편차 예측치 계산, 현재 영화의 평점과 각 사용자의 평균과의 차이(평점 편차)구하기
        if sim_scores.sum() > 0.0000001:
            prediction = np.dot(sim_scores, item_ratings) / sim_scores.sum() + 4.4   # 평균에서 얼마나 떨어져 있는가에 대한 예측
        else:
            prediction= 4.4
    else:
        prediction = 4.4                                             # rating 평균 넣기
    return prediction

score(ubcf_bias)
```

## UBCF_bias_knn

```python
def score_knn(model, neighbor_size=20):  # knn
    id_pairs = zip(x_test['user_id'], x_test['item_id'])
    y_pred = np.array([model(user, item, neighbor_size) for (user, item) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

```python
# KNN + 사용자 평가 경향 추천 (ubcf_bias_knn1)
def ubcf_bias_knn(user_id, item_id, neighbor_size=20):
    import numpy as np
    # 현 user의 평균 가져오기
    user_mean = rating_mean[user_id]
    if item_id in rating_matrix:
        sim_scores = user_similarity[user_id]                           # 현 user와 다른 사용자 간의 유사도 가져오기
        item_ratings = rating_matrix[item_id]                           # 현 movie에 대한 모든 사용자의 rating 가져오기
        others_mean = rating_mean                                       # 모든 사용자의 rating 평균 가져오기
        none_rating_idx = item_ratings[item_ratings.isnull()].index     # 현 movie에 대해 평가하지 않은 user 삭제
        item_ratings = item_ratings.drop(none_rating_idx)
        sim_scores = sim_scores.drop(none_rating_idx)
        others_mean = others_mean.drop(none_rating_idx)
        if neighbor_size == 0:                                          # Neighbor size가 지정되지 않은 경우
            item_ratings = item_ratings - others_mean                   # 영화 평점과 각 사용자의 평균 차이(편차) 예측치 계산
            prediction = np.dot(sim_scores, item_ratings) / sim_scores.sum()
            prediction = prediction + user_mean                         # 예측값에 현 사용자의 평균 더하기
        else:                                                           # Neighbor size가 지정된 경우
            neighbor_size = min(neighbor_size, len(sim_scores))         # 지정된 neighbor size 값과 해당 영화를 평가한 총사용자 수 중 작은 것으로 결정
            sim_scores = np.array(sim_scores)                           # array로 바꾸기 (argsort를 사용하기 위함)
            item_ratings = np.array(item_ratings)
            others_mean = np.array(others_mean)
            user_idx = np.argsort(sim_scores)                           # 유사도를 순서대로 정렬
            sim_scores = sim_scores[user_idx][-neighbor_size:]          # 유사도와 rating을 neighbor size만큼 받기
            item_ratings = item_ratings[user_idx][-neighbor_size:]
            others_mean = others_mean[user_idx][-neighbor_size:]        # 사용자의 mean을 neighbor size만큼 받기
            item_ratings = item_ratings - others_mean                   # 편차로 예측치 계산
            if sim_scores.sum() > 0.0000001:
                prediction = np.dot(sim_scores, item_ratings) / sim_scores.sum() + user_mean
            else:
                prediction= user_mean
    else:
        prediction = user_mean
    return prediction

score_knn(ubcf_bias_knn, 45)
```
