```python
import pandas as pd
import numpy as np

# Read rating data
ratings = pd.read_csv('/home/sjkim/recommendSystem/finalproject/Amazon_ratings.csv', encoding='latin-1')
ratings.drop('timestamp', axis=1, inplace=True)
ratings
```

```python
# Rating 데이터를 test, train으로 나누고 train을 full matrix로 변환
from sklearn.model_selection import train_test_split
x = ratings.copy()
y = ratings['user_id']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=12) # stratify=y 제거
```

```python
# UBCF와 다름!!! 실수하지 말 것!!! index=movie고 columns가 user임(1633x943)
rating_matrix_t = x_train.pivot(values='rating', index='item_id', columns='user_id')

# 아이템 pair의 Cosine similarities(유사도) 계산
from sklearn.metrics.pairwise import cosine_similarity
matrix_dummy = rating_matrix_t.copy().fillna(0)
item_similarity = cosine_similarity(matrix_dummy, matrix_dummy)     # 아이템 간 코사인 유사도를 item_similarity에 저장. 아이템간의 유사도 행렬
item_similarity = pd.DataFrame(item_similarity, index=rating_matrix_t.index, columns=rating_matrix_t.index)     # 계산이 편리하도록 DataFrame형식으로 바꿔줌
```

```python
# RMSE 계산을 위한 함수
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

def score(model):
    id_pairs = zip(x_test['user_id'], x_test['item_id'])
    y_pred = np.array([model(user, item) for (user, item) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

```python
def ibcf(user_id, item_id):                              # 하나의 user_id와 item_id 받아서 해당 사용자의 해당 영화에 대한 예측치를 계산하는 함수
    import numpy as np
    if user_id in rating_matrix_t:                        # 사용자(user_id)가 train set에 있는지 확인
        if item_id in item_similarity:                    # 현재 영화(item_id)가 train set에 있는지 확인
            sim_scores = item_similarity[item_id]         # 현재 영화와 다른 모든 영화와의 similarity 값 가져오기 (1633개)
            user_rating = rating_matrix_t[user_id]        # 현재 사용자의 모든 영화에 대한 평점 가져오기 (1633개)
            non_rating_idx = user_rating[user_rating.isnull()].index    # 현 사용자가 평가하지 않은 영화 index 가져오기 (rating 없는 거 지우게)
            user_rating = user_rating.dropna()            # 현 사용자가 평가하지 않은 영화 제거
            sim_scores = sim_scores.drop(non_rating_idx)  # 현 사용자가 평가하지 않은 영화의 similarity 값도 제거
            if sim_scores.sum() > 0.0000001:
                mean_rating = np.dot(sim_scores, user_rating) / sim_scores.sum()
            else:
                mean_rating= 4.4
        else:
            mean_rating = 4.4
    else:
        mean_rating = 4.4
    return mean_rating

score(ibcf)
```
